```python
from collections import OrderedDict # fastest way to remove duplicates from a list
import hashlib # inbuilt library containing hash functions 
import textwrap # used to split a string into group of n characters
```

```python
passwords_list = []

# Importing the file content into a list
with open("demo.txt", "r", encoding='cp437') as rockyou_file:
    
    # Iterate over each record (password)
    for password in rockyou_file:
        
        # Ignore empty records and those with length below 6
        password = password.strip()
        if password and len(password) > 7:
            passwords_list.append(password)
            
    # Remove spaces around the password, and remove the line breaker      
    for index, password in enumerate(passwords_list):
        passwords_list[index] = password.strip().replace('\n', '')


print(passwords_list)
```

Demonstration of each method

```python
password = '123456789'
```

```python
# Demonstration of hex digest generated by SHA256
password_digest = hashlib.sha256(password.encode('utf-8')).hexdigest()
print(password_digest)
print(len(password_digest))
print(type(password_digest))
```

```python
# Demonstration of splitting the SHA256 encrypted digest into 16 chunks, each of 4 hex characters
password_chunks = textwrap.wrap(password_digest, 4)
print(password_chunks)
```

```python
# Demonstration for converting each hex chunk into a decimal number
for chunk in password_chunks:
    equivalent_decimal = int(chunk, 16)
    print(chunk + " --> " + str(equivalent_decimal))
```

```python
# Demonstration of converting it into a csv
bitset_indices = []
isFirst = True
for chunk in password_chunks:
    equivalent_decimal = int(chunk, 16)
    bitset_indices.append(equivalent_decimal)
```

```python
# Sorting the indices in ascending order
bitset_indices.sort()
print(bitset_indices)

# Converting the sorted indices into a csv
sorted_indices = ",".join(map(str,bitset_indices))
print(type(sorted_indices))
print(sorted_indices) # Use this sorted_indices as the dictionary key
```

```python
# Overall operation

d = dict()

duplicate_count = 0

for password in passwords_list:
    
    password_digest = hashlib.sha256(password.encode('utf-8')).hexdigest()
    
    password_chunks = textwrap.wrap(password_digest, 4)
    
    bitset_indices = []
    
    isFirstChunk = True
    
    for chunk in password_chunks:
        
        equivalent_decimal = int(chunk, 16)
        
        bitset_indices.append(equivalent_decimal)

    bitset_indices.sort()
    
    decimal_csv = ",".join(map(str,bitset_indices))
    
    if decimal_csv not in d.keys():
        d[decimal_csv] = 1
    else:
        d[decimal_csv] += 1
        duplicate_count += 1
```
